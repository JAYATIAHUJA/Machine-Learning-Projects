```python
data_1 = data[["Car_Owner" , "Propert_Owner" ,"Annual_income" , "EDUCATION" , "label"]]
```

```python
from sklearn import preprocessing
```

```python
labelencoder = preprocessing.LabelEncoder()
```

```python
data_1["Car_Owner"] = labelencoder.fit_transform(data_1["Car_Owner"])
data_1["Propert_Owner"] = labelencoder.fit_transform(data_1["Propert_Owner"])
data_1["label"] = labelencoder.fit_transform(data_1["label"])
data_1["EDUCATION"] = labelencoder.fit_transform(data_1["EDUCATION"])
```

```python
data_1
```

```python
X = data_1.iloc[: , :-1]
Y = data_1.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2 )
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
def modelperformance(predictions):
    print("Accuracy score on model is {}".format(accuracy_score(Y_test,predictions)))
```

```python
from sklearn.linear_model import LogisticRegression
log_model = LogisticRegression()
log_model.fit(scaled_X_train , Y_train)
```

```python
log_predictions = log_model.predict(scaled_X_test)
```

```python
modelperformance(log_predictions)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.neighbors import KNeighborsClassifier
param_grid = {"n_neighbors" : [3,5,7,9,11],
              "weights" : ["uniform" , "distance"],
              "algorithm" : ["auto" , "kd_tree"]
              }
gridkn = GridSearchCV(  KNeighborsClassifier() , param_grid , cv =2 )
```

```python
gridkn.fit(scaled_X_train , Y_train)
```

```python
gridkn.best_params_
```

```python
knpreds = gridkn.predict(scaled_X_test)
```

```python
modelperformance(knpreds)
```

```python
from sklearn.svm import SVC
svm = SVC()
param_grid_svr = {"C" : [0.01 , 0.1 , 0.5 , 1] , 
                  "kernel": ["linear" , "rbf" , "poly"]}

gridsvr = GridSearchCV(svm,param_grid_svr)
```

```python
gridsvr.fit(scaled_X_train, Y_train)
```

```python
pred_svc = gridsvr.predict(scaled_X_test)
modelperformance(pred_svc)
```
